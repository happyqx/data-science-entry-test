```python
def string_reverse(s):
    """
    Task 1
    - Create a function that reverses a given string (s).
    - s must be a string.
    - Return the reversed string.
    """
    if not isinstance(s, str):
        raise TypeError("s must be a string")
    
    return s[::-1] # slicing technique to reverse the string -> sequence[start:stop:step]
```

```python
# Task 2
# Invoke the function "string_reverse" using the following scenarios:
# - "Hello World"
string_reverse("Hello World") # This should return "dlroW olleH"
```

```python
# - "Python"
string_reverse("Python")      # This should return "nohtyP"
```
