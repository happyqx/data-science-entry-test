```python
def update_dictionary(dct, key, value):
    """
    Task 1
    - Create a function that updates a dictionary (dct) with a new key-value pair.
    - If the key already exists in dct, print the original value, then update its value.
    - Return the updated dictionary.
    """
    if not isinstance(dct, dict):
        raise TypeError("dct must be a dictionary")

    if key in dct:
        print(f"Original value: {dct[key]}")

    dct[key] = value
    
    return dct
```

```python
# Task 2
# Invoke the function "update_dictionary" using the following scenarios:
# - {}, "name", "Alice"
update_dictionary({}, "name", "Alice") # This should return {'name': 'Alice'}
```

```python
# - {'age': 25}, "age", 26
update_dictionary({"age": 25}, "age", 26) # This should print "Original value: 25" and return {'age': 26}
```
