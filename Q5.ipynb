```python
import numbers
```

```python
def check_divisibility(num, divisor):
    """
    Task 1
    - Create a function to check if the number (num) is divisible by another number (divisor).
    - Both num and divisor must be numeric.
    - Return True if num is divisible by divisor, False otherwise.
    """
    if not (isinstance(num, numbers.Number) and isinstance(divisor, numbers.Number)):
        raise TypeError("Both num and divisor must be numeric")
    
    if divisor == 0:
        raise ValueError("divisor cannot be zero")
    
    quotient = num / divisor

    # Real numbers
    if isinstance(quotient, numbers.Real):
        return quotient.is_integer()  
    
    # Complex numbers
    elif isinstance(quotient, numbers.Complex):
        return quotient.real.is_integer() and quotient.imag.is_integer()
```

```python
# Task 2
# Invoke the function "check_divisibility" using the following scenarios:
# - 10, 2
check_divisibility(10, 2)  # This should return True
```

```python
# - 7, 3
check_divisibility(7, 3)   # This should return False
```
